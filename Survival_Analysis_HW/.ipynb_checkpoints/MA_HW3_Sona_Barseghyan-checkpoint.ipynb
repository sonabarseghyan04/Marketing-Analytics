```python
#importing neccessary libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from lifelines import LogNormalAFTFitter


from lifelines import (
    WeibullAFTFitter,
    LogNormalAFTFitter,
    LogLogisticAFTFitter,
    GeneralizedGammaRegressionFitter,
    KaplanMeierFitter
)
from lifelines.utils import concordance_index

from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.model_selection import train_test_split


try:
    plt.style.use('seaborn-v0_8-whitegrid')
except:
    plt.style.use('ggplot')
```

```python
#reading and exploring the dataset

filepath = "telco.csv"
df = pd.read_csv(filepath) 

df.info()
df.head()
df.describe(include='all')
```

```python
#making churn column binary
df['churn'] = df['churn'].map({ 'Yes':1, 'No':0 }).astype(int)  
```

```python
df.head()
```

```python
features = ['region','age','marital','address','income','ed','retire','gender','voice','internet','forward','custcat']

#seeing which features are categorical, to do one-hot inconding on them
numeric_cols = [i for i in features if df[i].dtype.kind in 'biufc' ]
cat_cols = [i for i in features if i not in numeric_cols]

df_enc = pd.get_dummies(df[features], columns=cat_cols, drop_first=True)

#merging with event and duration(churn and tenure)
df_for_model = pd.concat([df[['ID','tenure','churn']].reset_index(drop=True), df_enc.reset_index(drop=True)], axis=1)
df_for_model = df_for_model.set_index('ID')
```

```python
#standardizing numeric features
scaler = StandardScaler()
if numeric_cols:
    df_for_model[numeric_cols] = scaler.fit_transform(df_for_model[numeric_cols])
```

```python
#splitting data
train_df, test_df = train_test_split(df_for_model, test_size=0.25, random_state=42, stratify=df_for_model['churn'])
```

```python
duration = 'tenure'
event = 'churn'

X_train = train_df.drop(columns=[duration, event])
train_durations = train_df[duration]
train_events = train_df[event]

X_test = test_df.drop(columns=[duration, event])
test_durations = test_df[duration]
test_events = test_df[event]
```

```python
#fitting models

weibull = WeibullAFTFitter()
lognormal = LogNormalAFTFitter()
loglogistic = LogLogisticAFTFitter()
gg = GeneralizedGammaRegressionFitter(penalizer=0.01)

models = {
    'WeibullAFT': weibull,
    'LogNormalAFT': lognormal,
    'LogLogisticAFT': loglogistic,
    'GeneralizedGamma': gg
}

fitted_models = {}
for name, model in models.items():
    print(f"Fitting {name} ...")
    try:
        model.fit(train_df, duration_col=duration, event_col=event)
        fitted_models[name] = model
        print(name, "fitted.")
    except Exception as e:
        print(f"Failed to fit {name}: {e}")
```

### Generalized Gamma AFT failed to converge due to high model complexity and multicollinearity. The model was excluded from the comparison.

```python
#model comparison

results = []

for name, model in fitted_models.items():
    try:
        ci = model.score(train_df, scoring_method="concordance_index")
    except:
        ci = None

    results.append({
        "Model": name,
        "AIC": model.AIC_,
        "Log-likelihood": model.log_likelihood_,
        "Concordance index": ci
    })

df_results = pd.DataFrame(results)
df_results.sort_values("AIC", inplace=True)

print("\nModel comparison:")
print(df_results)
```

```python
example_customer = train_df.drop(columns=[duration, event]).iloc[0:1]

plt.figure(figsize=(8,6))

for name, model in fitted_models.items():
    if name == "GeneralizedGamma":
        continue 
    surv = model.predict_survival_function(example_customer)
    plt.step(surv.index, surv.values.flatten(), where='post', label=name)

plt.title("Predicted Survival Curves")
plt.xlabel("Tenure (months)")
plt.ylabel("Survival Probability")
plt.legend()
plt.show()
```

I would choose the LogNormalAFT model because it provides a good balance between accuracy and interpretability. It fits the data well, showing a realistic pattern of churn where survival drops faster at the beginning and then stabilizes. The model converges reliably, unlike the GeneralizedGamma, and its coefficients are easy to interpret, helping identify which factors increase or decrease churn risk. It also produces survival probabilities that can be directly used to calculate Customer Lifetime Value (CLV) and guide retention strategies.  

```python
final_model = fitted_models['LogNormalAFT']
```

```python
alpha = 0.05

lognormal_summary = final_model.summary

significant_features = lognormal_summary[lognormal_summary['p'] < alpha].index.tolist()
print("Significant features:")
print(significant_features)

significant_cols = []
for feature in significant_features:
    if isinstance(feature, tuple):
        col_name = feature[1]
        if '_' in col_name:
            base_col = col_name.split('_')[0]
            if base_col == 'custcat':
                significant_cols.append('custcat')
            else:
                significant_cols.append(base_col)
        else:
            significant_cols.append(col_name)
    else:
        if '_' in feature:
            significant_cols.append(feature.split('_')[0])
        else:
            significant_cols.append(feature)


significant_cols = list(set(significant_cols))
print("Significant columns to keep:", significant_cols)
```

```python
final_features = ['tenure', 'churn_event'] + significant_cols

final_features = [col for col in final_features if col in df.columns]

final_data = pd.get_dummies(
    df[final_features],
    columns=[col for col in significant_cols if col in cat_cols],
    drop_first=True)

final_data['churn_event'] = df['churn']
```

```python
final_model_signif = LogNormalAFTFitter()
final_model_signif.fit(final_data, duration_col='tenure', event_col='churn_event')
final_model_signif.summary
```

```python
def calculate_clv(survival_prob, avg_monthly_revenue, discount_rate=0.01):
    discount_factors = np.array([(1 / (1 + discount_rate)) ** t for t in range(len(survival_prob))])
    return avg_monthly_revenue * np.sum(survival_prob * discount_factors)

avg_monthly_revenue = 60
timeline = np.arange(1, 37)

clv_values = []
for _, row in final_data.iterrows():
    customer_data = pd.DataFrame([row])
    survival_prob = final_model_signif.predict_survival_function(customer_data, times=timeline).values.flatten()
    clv_values.append(calculate_clv(survival_prob, avg_monthly_revenue))

df['CLV'] = clv_values

df.head()
```

```python
print("CLV Summary Statistics:")
print(df['CLV'].describe())

plt.figure(figsize=(12, 6))
sns.histplot(data=df, x='CLV', bins=30, kde=True)
plt.title('Customer Lifetime Value (CLV) Distribution')
plt.xlabel('CLV ($)')
plt.show() 
```

```python
# Average CLV by segment 
segment_clv = df.groupby(['custcat', 'voice', 'internet'])['CLV'].agg(['mean', 'count']).reset_index()
segment_clv = segment_clv.sort_values('mean', ascending=False)
print("Top 10 Customer Segments by CLV:")
print(segment_clv.head(10))

#12-month churn probability
timeline_12m = np.array([12])
churn_prob_12m = []

for _, row in final_data.iterrows():
    customer_data = pd.DataFrame([row])
    survival_prob = final_model_signif.predict_survival_function(customer_data, times=timeline_12m).values.flatten()[0]
    churn_prob = 1 - survival_prob
    churn_prob_12m.append(churn_prob)

df['churn_prob_12m'] = churn_prob_12m

#high-value and high-risk customers
high_value_threshold = df['CLV'].quantile(0.75)
high_risk_threshold = df['churn_prob_12m'].quantile(0.75)

df['high_value'] = df['CLV'] >= high_value_threshold
df['high_risk'] = df['churn_prob_12m'] >= high_risk_threshold
df['high_value_at_risk'] = df['high_value'] & df['high_risk']


print("High-value customers:", df['high_value'].sum())
print("High-risk customers:", df['high_risk'].sum())
print("High-value at-risk customers:", df['high_value_at_risk'].sum())

# Potential revenue loss
if df['high_value_at_risk'].sum() > 0:
    annual_revenue_loss = df[df['high_value_at_risk']]['CLV'].sum() * df[df['high_value_at_risk']]['churn_prob_12m'].mean()
else:
    annual_revenue_loss = df[df['high_value']]['CLV'].sum() * df['churn_prob_12m'].mean()
    
print(f"Potential annual revenue loss from at-risk valuable customers: ${annual_revenue_loss:.2f}")

#retention budget 
retention_budget = annual_revenue_loss * 0.15
print(f"Suggested annual retention budget: ${retention_budget:.2f}")


segment_analysis = df.groupby('custcat').agg({
    'CLV': 'mean',
    'churn_prob_12m': 'mean',
    'ID': 'count'
}).rename(columns={'ID': 'count'}).reset_index()

print("\nCustomer Segment Analysis:")
print(segment_analysis)
```
